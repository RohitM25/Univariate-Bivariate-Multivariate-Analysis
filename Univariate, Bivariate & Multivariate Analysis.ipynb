# Univariate, Bivariate & Multivariate Analysis


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

### Iris Dataset

```python
df= pd.read_csv('https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv')
```

```python
df.head(10)
```

### Univariate analysis

```python
df_setosa = df.loc[df['species']=='setosa']
```

```python
df_virginica = df.loc[df['species']=='virginica']
```

```python
df_versicolor = df.loc[df['species']=='versicolor']
```

```python
plt.plot(df_setosa['sepal_length'],np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'],np.zeros_like(df_virginica['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'],np.zeros_like(df_versicolor['sepal_length']),'o')
plt.show()
```

### Bivariate analysis

```python
sns.FacetGrid(df,hue = 'species',size = 5).map(plt.scatter,'sepal_length','petal_length').add_legend()
plt.show()
```

Observation: Setosa can be easily obtained but other two features have overlapping points

### Multivariate Analysis

```python
sns.pairplot(df,hue='species',size = 5).add_legend()
```
